```python
X=df[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
y=df.Price
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(random_state=1)
model.fit(X,y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices=model.predict(X)
mean_absolute_error(y,predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
melbourne_model=DecisionTreeRegressor()
melbourne_model.fit(X_train,y_train)
test_predictions=melbourne_model.predict(X_test)
print(mean_absolute_error(test_predictions,y_test))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
def get_mae(max_leaf_nodes,X_train,X_test,y_train,y_test):
  model=DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state=0)
  model.fit(X_train,y_train)
  mae=mean_absolute_error(y_test,model.predict(X_test))
  return mae
```

```python
mae_results = {}
for max_leaf_nodes in [5,50,500,5000]:
  my_mae=get_mae(max_leaf_nodes,X_train,X_test,y_train,y_test)
  mae_results[max_leaf_nodes] = my_mae
  print('max_leaf_nodes:%d \t\t,my_mae:%d'%(max_leaf_nodes,my_mae))
```

```python
best_tree_size=min(mae_results,key=mae_results.get)
```

```python
final_model=DecisionTreeRegressor(max_leaf_nodes=best_tree_size,random_state=1)
final_model.fit(X,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
forest_model=RandomForestRegressor(random_state=1)
forest_model.fit(X_train,y_train)
print('MAE:', mean_absolute_error(y_test,forest_model.predict(X_test)))
```

```python
forest1=RandomForestRegressor()
forest1.fit(X,y)
```

```python
test_data_path = 'melb_data.csv'
test_data = pd.read_csv(test_data_path)
# Drop rows with missing values in the relevant columns, consistent with training data preprocessing
test_data = test_data.dropna(axis=0)
X_test=test_data[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
test_predictions=forest1.predict(X_test)
print(test_predictions)
```

```python
output = pd.DataFrame({'Id': test_data.index,
                       'HousePrice': test_predictions})
output.to_csv('submission.csv', index=False)
```
